```python
fig, ax = plt.subplots(figsize=(10,7))
ShipModesum=df.loc[:,['Ship Mode','Profit']]
ShipModesum=ShipModesum.groupby(['Ship Mode'], as_index=False).sum()
sns.barplot(x=ShipModesum['Ship Mode'], y=ShipModesum['Profit'])
plt.savefig('Shipmode_sum.png')
```

```python
ShipModesum
```

Here we can see that **Standard Ship Mode** has both the **MAX** and **MINIMUM** value and is also the biggest contributor in Profits.


Now let us look at the **Category** feature. As seen in data.head() there are some categories of items to be sold in supermarkets.  
We will follow the same steps as we followed while gaining insights from 'Ship Mode' feature.    
**Step1. What are the Unique subcategories it has.**  
**Step2. What is representation of profitandloss with each subcategory.**  
**Step3. How much each category contribute to the total profit.**  

```python
df['Category'].unique()
```

```python
sns.boxplot(df['Category'],profitandloss)
plt.savefig('Category_Boxplot.png')
```

Here we can see that **Technology** has the both the **Extreme** Points w.r.t Profits.  

```python
fig, ax = plt.subplots(figsize=(9,7))
sns.countplot(x= df['Category'])
plt.savefig('Category_Countplot.png')
```

The one thing to note here is that although Technology has the max profit point but the Office Supplies items were sold the most.

```python
fig, ax = plt.subplots(figsize=(10,7))
Categorysum=df.loc[:,['Category','Profit']]
Categorysum=Categorysum.groupby(['Category'], as_index=False).sum()
sns.barplot(x=Categorysum['Category'], y=Categorysum['Profit'])
plt.savefig('Category_sum.png')
```

**Although Office supplies were sold the most , Technology category still contribute the most to the total profit.  
Despite that Office supplies seems to be a stable source as it is being sold the most and make decent contribution to it.  
Furniture does not show good performance this is the area we need to work upon**

Let us take a look at the **Sub-Category** column for a deeper insight as what is causing such **Bad Performance** of **Furniture** Category.

```python
category_wise_profit=df.loc[:,['Category', 'Sub-Category', 'Sales','Quantity', 'Profit','Discount']]
category_wise_profit.head()
```

```python
furniture_profits=category_wise_profit[category_wise_profit['Category']=='Furniture']
furniture_profits.groupby(['Sub-Category'], as_index=False).sum()
```

Here we can see that **Bookcases** and **Tables** are the reason why **Furniture** Column's is not showing great returns.  
We should invest more in **Chairs** as it adds the maximum profit in Furniture column.

**Now let us analyse the performance of supermarkets according to their locations.**

```python
locations_wise_profits=df.loc[:,['State', 'City', 'Sales', 'Quantity', 'Profit']]
state_profits = locations_wise_profits.groupby(['State'], as_index=False).sum()
state_profits
```

**Lets look how many stores are in loss according to their States.**

```python
State_with_loss = state_profits[state_profits['Profit']<=0]
State_with_loss
```

```python
fig, ax = plt.subplots(figsize=(15,7))
sns.barplot(y=State_with_loss['Profit'],x=State_with_loss['State'])
```

Here we can see that our super store at **Texas** is facing biggest loses despite being the most number of quantities sold. This Superstore needs to be checked and improved upon. 

```python
Items_and_positions = df.loc[:,['State','City','Category','Sub-Category','Quantity']]
Texas_sell = Items_and_positions[Items_and_positions['State']=='Texas']
sns.barplot(x=Texas_sell['Category'],y=Texas_sell['Quantity'])
plt.savefig('Texas_items.png')
```

As we can see, Texas sells less Technology Category items which gives high returns whereas it sells Furniture( which is a loosing category) in abundance.
